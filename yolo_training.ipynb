```python
# First working model
import torch
import cv2
import pytesseract
import pandas as pd
from datetime import datetime
import pytz
import os
from pathlib import Path
from google.colab import files

# Setup directories
snapshot_dir = Path("/content/snapshots")
snapshot_dir.mkdir(exist_ok=True)
log_dir = Path("/content/logs")
log_dir.mkdir(exist_ok=True)

# Get IST Timezone
ist = pytz.timezone('Asia/Kolkata')
today_str = datetime.now(ist).strftime('%d_%m_%Y')
csv_path = f"/content/logs/dip_log_{today_str}.csv"

# Counter file to persist daily counter
counter_file = f"/content/logs/counter_{today_str}.txt"
counter = 1
if os.path.exists(counter_file):
    with open(counter_file, 'r') as f:
        counter = int(f.read())

# Load YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'custom',
                       path='/content/drive/MyDrive/Project/yolo_dataset/yolov5/runs/train/exp/weights/best.pt')
model.conf = 0.25

# Input/output video paths
input_video = '/content/drive/MyDrive/Project/yolo_dataset/vsdc-sr 2025-07-23 13-47-36.mp4'
output_video = '/content/yolo_dip_in_out_logged.mp4'

cap = cv2.VideoCapture(input_video)
width, height = int(cap.get(3)), int(cap.get(4))
fps = cap.get(cv2.CAP_PROP_FPS)
out = cv2.VideoWriter(output_video, cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

# ROI (bottom-center, narrow)
roi_x1, roi_y1 = int(width * 0.2), int(height * 0.8)
roi_x2, roi_y2 = int(width * 0.7), int(height * 1.0)

track_classes = ["car", "number_plate"]
dip_status = "Outside"
in_counter = 0
out_counter = 0
frame_num = 0
log = []

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)
    detections = results.pandas().xyxy[0]
    inside = False
    plate_text = ""

    for _, row in detections.iterrows():
        label = row['name']
        conf = row['confidence']
        xmin, ymin, xmax, ymax = map(int, [row['xmin'], row['ymin'], row['xmax'], row['ymax']])
        cv2.rectangle(frame, (xmin, ymin), (xmax, ymax), (0, 255, 0), 2)
        cv2.putText(frame, f'{label} {conf:.2f}', (xmin, ymin - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

        if label in track_classes:
            center_x = (xmin + xmax) // 2
            center_y = (ymin + ymax) // 2
            if roi_x1 <= center_x <= roi_x2 and roi_y1 <= center_y <= roi_y2:
                inside = True

            if label == "number_plate" and roi_y1 <= center_y <= roi_y2:
                cropped = frame[ymin:ymax, xmin:xmax]
                try:
                    plate_text = pytesseract.image_to_string(cropped, config='--psm 8')
                    plate_text = plate_text.strip().replace('\n', '').replace('\x0c', '')
                except:
                    plate_text = ""

    # Time
    timestamp = datetime.now(ist).strftime('%Y-%m-%d %H:%M:%S')

    # Dip logic
    if inside:
        in_counter += 1
        out_counter = 0
    else:
        out_counter += 1
        in_counter = 0

    # Dip In
    if in_counter >= 4 and dip_status != "Inside":
        dip_status = "Inside"
        snap_path = snapshot_dir / f"dip_in_{counter}.jpg"
        cv2.imwrite(str(snap_path), frame)
        log.append([counter, "Dip In", timestamp, plate_text, str(snap_path)])
        counter += 1

    # Dip Out
    if out_counter >= 2 and dip_status != "Outside":
        dip_status = "Outside"
        snap_path = snapshot_dir / f"dip_out_{counter}.jpg"
        cv2.imwrite(str(snap_path), frame)
        log.append([counter, "Dip Out", timestamp, plate_text, str(snap_path)])
        counter += 1

    # Draw ROI and status
    cv2.rectangle(frame, (roi_x1, roi_y1), (roi_x2, roi_y2), (255, 0, 0), 2)
    cv2.putText(frame, 'ROI Zone', (roi_x1 + 10, roi_y1 - 10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 0, 0), 2)
    cv2.putText(frame, f'Status: {dip_status}', (20, 40),
                cv2.FONT_HERSHEY_SIMPLEX, 1,
                (0, 0, 255) if dip_status == "Inside" else (100, 100, 100), 2)

    out.write(frame)
    frame_num += 1

cap.release()
out.release()

# Save updated counter
with open(counter_file, 'w') as f:
    f.write(str(counter))

# Ask user to verify/edit PlateText for each record
for i in range(len(log)):
    plate = log[i][3]
    print(f"🔎 Detected plate: {plate}")
    user_input = input("Do you want to edit it? (press Enter to keep, or type new plate): ").strip()
    if user_input:
        log[i][3] = user_input

# Parse log entries and check duration
in_time = None
violation_log = []

for row in log:
    _, event, ts_str, _, _ = row
    ts = datetime.strptime(ts_str, "%Y-%m-%d %H:%M:%S")

    if event == "Dip In":
        in_time = ts
        in_row = row.copy()
    elif event == "Dip Out" and in_time:
        duration = (ts - in_time).total_seconds()
        status = f"Violation: Stayed {duration:.1f} seconds" if duration >= 4 else f"No Violation: Stayed {duration:.1f} seconds"
        in_row.append(status)
        row.append(status)
        violation_log.extend([in_row, row])
        in_time = None

# Save CSV
new_df = pd.DataFrame(violation_log, columns=["Counter", "Event", "Timestamp", "PlateText", "Snapshot", "ViolationStatus"])
if os.path.exists(csv_path):
    old_df = pd.read_csv(csv_path)
    combined = pd.concat([old_df, new_df], ignore_index=True)
else:
    combined = new_df

combined.to_csv(csv_path, index=False)
print("✅ All saved!")

# Downloads
files.download(output_video)
files.download(csv_path)
```
